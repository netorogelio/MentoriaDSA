```python
# Baixar o chromedriver.exe criar uma pasta e mover o arquivo para dentro dela
# Baixar o arquivo chromedrive.exe conforme a versão do seu navegaror
# !python -m selenium
# ============================================================================
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import shutil
import pandas as pd

url = 'http://www.educacao.sp.gov.br/central-de-atendimento/downloads.asp'
drive = webdriver.Chrome(executable_path=r'C:/Mentoria/Dataset/chromedriver.exe')
time.sleep(3)
drive.get(url)
time.sleep(3)
texto = '//*[@id="content"]/article/section/p[1]/a[sequencia]'
# Baixar os arquivos das escolas do estado de São Paulo
# ESTADUAL_SEE.xlsx
# ESTADUAL_OUTRAS.xlsx
# FEDERAL.xlsx
# MUNICIPAL.xlsx
# PRIVADA.xlsx
# ======================================================
arquivo = ['ESTADUAL_SEE.xlsx','ESTADUAL_OUTRAS.xlsx','FEDERAL.xlsx','MUNICIPAL.xlsx','PRIVADA.xlsx']

for i in range(1,len(arquivo)+1):
    texto1 = texto.replace('sequencia', str(i))
    if i >= 2:
        texto1 = texto1 + '/b'
    drive.find_element_by_xpath(texto1).click()
    time.sleep(4)
drive.close()    
# Mover os arquivos para o diritorio de trabalho  
for i in arquivo:
    origem = r'C://Users//DELL//Downloads//' + i
    destino = r'C://Mentoria//Dataset//Temporario//' + i 
    shutil.move(origem,destino)
```

```python
df_estadual_see = pd.read_excel(r'C:\Mentoria\Dataset\Temporario\ESTADUAL_SEE.xlsx', skiprows=5) #,index_col=0)
df_estadual_outras = pd.read_excel(r'C:\Mentoria\Dataset\Temporario\ESTADUAL_OUTRAS.xlsx', skiprows=5) #,index_col=0)
df_federal = pd.read_excel(r'C:\Mentoria\Dataset\Temporario\FEDERAL.xlsx', skiprows=5) #,index_col=0)
df_municipal = pd.read_excel(r'C:\Mentoria\Dataset\Temporario\MUNICIPAL.xlsx', skiprows=5) #,index_col=0)
df_privada = pd.read_excel(r'C:\Mentoria\Dataset\Temporario\PRIVADA.xlsx', skiprows=5) #,index_col=0)

df_estadual_see = df_estadual_see[['Nome dependencia Administrativa','Codigo Municipio','Municipio','CEP',
                                   'Creche','Pre','1º Ano Ef','2º Ano Ef','3º Ano Ef','4º Ano Ef',
                                   '5º Ano Ef','6º Ano Ef','7º Ano Ef','8º Ano Ef','9º Ano Ef','1ª Serie Em',
                                   '2ª Serie Em','3ª Serie Em']]

df_estadual_see
```

```python
df_estadual_outras = df_estadual_outras[['Nome dependencia Administrativa','Codigo Municipio','Municipio','CEP',
                                         'Creche','Pre','1º Ano Ef','2º Ano Ef','3º Ano Ef','4º Ano Ef',
                                         '5º Ano Ef','6º Ano Ef','7º Ano Ef','8º Ano Ef','9º Ano Ef','1ª Serie Em',
                                         '2ª Serie Em','3ª Serie Em']]
df_estadual_outras
```

```python
df_federal = df_federal[['Nome dependencia Administrativa','Codigo Municipio','Municipio','CEP',
                         'Creche','Pre','1º Ano Ef','2º Ano Ef','3º Ano Ef','4º Ano Ef',
                         '5º Ano Ef','6º Ano Ef','7º Ano Ef','8º Ano Ef','9º Ano Ef','1ª Serie Em',
                         '2ª Serie Em','3ª Serie Em']]
df_federal
```

```python
df_municipal = df_municipal[['Nome dependencia Administrativa','Codigo Municipio','Municipio','CEP',
                             'Creche','Pre','1º Ano Ef','2º Ano Ef','3º Ano Ef','4º Ano Ef',
                             '5º Ano Ef','6º Ano Ef','7º Ano Ef','8º Ano Ef','9º Ano Ef','1ª Serie Em',
                             '2ª Serie Em','3ª Serie Em']]
df_municipal
```

```python
df_privada = df_privada[['Nome dependencia Administrativa','Codigo Municipio','Municipio','CEP',
                         'Creche','Pre','1º Ano Ef','2º Ano Ef','3º Ano Ef','4º Ano Ef',
                         '5º Ano Ef','6º Ano Ef','7º Ano Ef','8º Ano Ef','9º Ano Ef','1ª Serie Em',
                         '2ª Serie Em','3ª Serie Em']]
df_privada
```

```python
df_geral = pd.concat([df_estadual_see, df_estadual_outras, df_federal, df_municipal,df_privada])
df_geral
```

```python
df_geral = df_geral.drop(columns=['CEP'])
df_geral
```

```python
df_geral_munic = df_geral
df_geral_munic = df_geral_munic.drop(columns=['Nome dependencia Administrativa'])
df_geral_munic = df_geral.groupby(by=['Codigo Municipio','Municipio']).sum()
ano = 2017
df_geral_munic.insert(0,"Ano",ano,True)
df_geral_munic
```

```python
df_geral_dep_adm = df_geral
df_geral_dep_adm = df_geral_dep_adm.drop(columns=['Codigo Municipio','Municipio'])
df_geral_dep_adm = df_geral_dep_adm.groupby(by=['Nome dependencia Administrativa']).sum()
ano = 2017
df_geral_dep_adm.insert(0,"Ano",ano,True)
df_geral_dep_adm
```

```python
df_geral = pd.concat([df_estadual_see, df_estadual_outras, df_federal, df_municipal,df_privada])
ano = 2017
df_geral.insert(0,"Ano",ano,True)
df_geral
```

```python
df_geral.to_excel(r"C:\Mentoria\Dataset\Temporario\ESCOLA_GERAL.xlsx")
df_geral_dep_adm.to_excel(r"C:\Mentoria\Dataset\Temporario\ESCOLA_DEP_ADM.xlsx")
df_geral_munic.to_excel(r"C:\Mentoria\Dataset\ESCOLA_MUNICIPIO.xlsx")
df_geral
```
