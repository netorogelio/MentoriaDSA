```python
# Baixar o chromedriver.exe criar uma pasta e mover o arquivo para dentro dela
# Baixar o arquivo chromedrive.exe conforme a versão do seu navegaror
# !python -m selenium
# ============================================================================
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import shutil
import pandas as pd

drive = webdriver.Chrome(executable_path=r'C:/Mentoria/Dataset/chromedriver.exe')
time.sleep(3)
# Baixar os arquivos das renda per capita do estado de São Paulo
# tab_pib_2018.xlsx
# tab_pib_2017.xlsx
# tab_pib_2016.xlsx
# tab_pib_2015.xlsx
# tab_pib_2014.xlsx
# tab_pib_2013.xlsx
# tab_pib_2012.xlsx
# tab_pib_2011.xlsx
# tab_pib_2010.xlsx
# tab_pib_2009.xlsx
# tab_pib_2008.xlsx
# tab_pib_2007.xlsx
# tab_pib_2006.xlsx
# tab_pib_2005.xlsx
# tab_pib_2004.xlsx
# tab_pib_2003.xlsx
# tab_pib_2002.xlsx
# ======================================================      
arquivo = ['tab_pib_2018.xlsx','tab_pib_2017.xlsx','tab_pib_2016.xlsx','tab_pib_2015.xlsx','tab_pib_2014.xlsx',
           'tab_pib_2013.xlsx','tab_pib_2012.xlsx','tab_pib_2011.xlsx','tab_pib_2010.xlsx','tab_pib_2009.xlsx',
           'tab_pib_2008.xlsx','tab_pib_2007.xlsx','tab_pib_2006.xlsx','tab_pib_2005.xlsx','tab_pib_2004.xlsx',
           'tab_pib_2003.xlsx','tab_pib_2002.xlsx']
url = '//*[@id="dataset-resources"]/ul/li[sequencia]/a'
for i in range(1,len(arquivo)+1):
    drive.get('http://repositorio.seade.gov.br/dataset/pib-municipal-2002-2018')
    time.sleep(4)    
    drive.find_element_by_xpath(url.replace('sequencia', str(i))).click()
    time.sleep(4)
    drive.find_element_by_xpath('//*[@id="content"]/div[3]/section/div/p/a').click()
    time.sleep(4)
drive.close()
# Mover os arquivos para o diritorio de trabalho
for i in arquivo:
    origem = r'C://Users//DELL//Downloads//' + i
    destino = r'C://Mentoria//Dataset//Temporario//' + i 
    shutil.move(origem,destino)
```

```python
arquivo = ['tab_pib_2018.xlsx','tab_pib_2017.xlsx',
           'tab_pib_2016.xlsx',
           'tab_pib_2015.xlsx',
           'tab_pib_2014.xlsx',
           'tab_pib_2013.xlsx','tab_pib_2012.xlsx','tab_pib_2011.xlsx','tab_pib_2010.xlsx','tab_pib_2009.xlsx',
           'tab_pib_2008.xlsx','tab_pib_2007.xlsx','tab_pib_2006.xlsx','tab_pib_2005.xlsx','tab_pib_2004.xlsx',
           'tab_pib_2003.xlsx','tab_pib_2002.xlsx']
df_geral = pd.DataFrame()
for i in arquivo:
#    print(i)
    if i == 'tab_pib_2015.xlsx':
        pula_linha = 9 
    else:
        pula_linha = 10
        
    ano = i[8:12]
    
    df_tab_pib = pd.read_excel(r'C://Mentoria//Dataset//Temporario//' + i, skiprows=pula_linha)  
    df_tab_pib = df_tab_pib.rename(columns={'Unnamed: 0':'Municípios', 
                                            'Unnamed: 1':'Agropecuária',
                                            'Unnamed: 2':'Indústria',
                                            'Unnamed: 3':'Administração Pública',
                                            'Unnamed: 4':'Total  (exclusive Administração Pública)',
                                            'Unnamed: 5':'Total geral',
                                            'Unnamed: 6':'Impostos',
                                            'Unnamed: 7':'PIB',
                                            'Unnamed: 8':'PIB per Capita'})

    df_tab_pib.insert(0,"Ano",ano,True)
    # Retirar as linhas de rodape do arquivo
    linha = len(df_tab_pib)-1
    for i in range(0,5):
        df_tab_pib = df_tab_pib.drop(linha)
        linha = linha-1

    # Gera o dataframe geral com todas as informações
    if df_geral.empty:
        df_geral = df_tab_pib
    else:
        df_geral = pd.concat([df_geral, df_tab_pib])
        
# df_geral
# Retirar as colunas geradas a mais
df_geral = df_geral.drop(columns=['Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12',
                                  'Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16',
                                  'Unnamed: 17','Unnamed: 18','Unnamed: 19'])
df_geral    
```

```python
df_geral = df_geral
df_geral.Municípios = df_geral.Municípios.str.upper()
df_geral.Municípios = df_geral.Municípios.str.replace('Á','A').str.replace('Í','I').str.replace('Ó','O')
df_geral.Municípios = df_geral.Municípios.str.replace('Ã','A').str.replace('É','E').str.replace('Â','A')
df_geral.Municípios = df_geral.Municípios.str.replace('Ê','E').str.replace('Ô','O').str.replace('Ç','C')
df_geral.Municípios = df_geral.Municípios.str.replace('Ú','U').str.replace('Õ','O')
df_geral.to_excel(r"C:\Mentoria\Dataset\Temporario\PIB_MUNICIAPAL.xlsx")
df_geral
```
