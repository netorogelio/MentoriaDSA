```python
# Baixar o chromedriver.exe criar uma pasta e mover o arquivo para dentro dela
# Baixar o arquivo chromedrive.exe conforme a versão do seu navegaror
# !python -m selenium
# ============================================================================
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import shutil
import pandas as pd

drive = webdriver.Chrome(executable_path=r'C:/Mentoria/Dataset/chromedriver.exe')
time.sleep(3)
drive.get('https://repositorio.seade.gov.br/dataset/populacao-por-municipio-sexo-e-idade')
time.sleep(4)    
drive.find_element_by_xpath('//*[@id="dataset-resources"]/ul/li[1]/a').click()
time.sleep(4)
drive.find_element_by_xpath('//*[@id="content"]/div[3]/section/div/p/a').click()
time.sleep(4)
drive.close()
origem = r'C://Users//DELL//Downloads//' + 'populacao_2021_sexoidade_esp.csv'
destino = r'C://Mentoria//Dataset//Temporario//' + 'populacao_2021_sexoidade_esp.csv'
shutil.move(origem,destino)
```

```python
df_geral = pd.read_csv("C:/Mentoria/Dataset/Temporario/populacao_2021_sexoidade_esp.csv", delimiter=";", encoding='ISO-8859-1')
df_geral.municipio = df_geral.municipio.str.upper()
df_geral.municipio = df_geral.municipio.str.replace('Á','A').str.replace('Í','I').str.replace('Ó','O')
df_geral.municipio = df_geral.municipio.str.replace('Ã','A').str.replace('É','E').str.replace('Â','A')
df_geral.municipio = df_geral.municipio.str.replace('Ê','E').str.replace('Ô','O').str.replace('Ç','C')
df_geral.municipio = df_geral.municipio.str.replace('Ú','U').str.replace('Õ','O')
df_geral.populacao = df_geral.populacao.str.replace('-   ','0').str.replace('.','')
df_geral.to_excel(r"C:\Mentoria\Dataset\POP_MUNICIPIO.xlsx")
df_geral
```

```python
PopMunicipio = pd.read_excel(r"C:\Mentoria\Dataset\POP_MUNICIPIO.xlsx",index_col=0)
PopMunicipio = PopMunicipio.drop(columns=['cod_ibge','faixa_etaria'])
PopMunMulher = PopMunicipio.query('sexo=="Mulheres"')
PopMunMulher = PopMunMulher.groupby(by=['municipio','sexo','ano']).sum()
PopMunMulher = PopMunMulher.rename(columns={'populacao':'mulher'})
PopMunMulher
```

```python
PopMunicipio = pd.read_excel(r"C:\Mentoria\Dataset\POP_MUNICIPIO.xlsx",index_col=0)
PopMunicipio = PopMunicipio.drop(columns=['cod_ibge','faixa_etaria'])
PopMunHomem = PopMunicipio.query('sexo=="Homens"')
PopMunHomem = PopMunHomem.groupby(by=['municipio','sexo','ano']).sum()
PopMunHomem = PopMunHomem.rename(columns={'populacao':'homem'})
PopMunHomem
```

```python
dataRes = pd.merge(PopMunMulher, PopMunHomem, how = 'left', on = ['municipio','ano'])
dataRes
dataRes.to_excel(r"C:\Mentoria\Dataset\POP_MUNICIPIO.xlsx")
```
